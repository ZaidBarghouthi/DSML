```python
import numpy as np
import matplotlib.pyplot as plt

x = np.arange(-55,50,0.1)
y1 = [d**2 for d in x]
y2 = [d**3 for d in x]
plt.plot(x,y1, label="y1")
plt.plot(x,y2, label="y2")
plt.axis([-50,50,-1000,1000])
plt.xlabel("Time (s)")
plt.ylabel("Value")
plt.grid(True)
plt.legend()
plt.title("My Fig")
plt.savefig('ss.png')
plt.show()
```
